```python
def write_csv_title(filename, titles):
    with open(filename, 'w', newline='', encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerow(titles)

def write_csv_elements(filename, elements):
    with open(filename, 'a+', newline='', encoding="utf-8") as csvfile:
        writer = csv.writer(csvfile, delimiter=',')
        writer.writerow(elements)
```

```python
def combine_all_files(path):
    import os
    name_list = os.listdir(path) 
    if ".ipynb_checkpoints" in name_list:
        name_list.remove(".ipynb_checkpoints")
    if "desktop.ini" in name_list:
        name_list.remove("desktop.ini")
    data = []
    for i in name_list:
        x = pd.read_csv(path + i).dropna(axis = 0, how = "all")
        data.append(x)
    data = pd.concat(data)
    data = data.reset_index().drop(columns = {"index"})
    return data
```

```python
def click_button(button_xpath, url, sleeptime_start, sleeptime_end):
    driver.get(url)
    while True:
        try:
            element = driver.find_element_by_xpath(button_xpath)
            element.click()
            time.sleep(random.randint(sleeptime_start, sleeptime_end))
        except NoSuchElementException:
            break
```

```python
def get_days_positions(date_in):
    import datetime
    import calendar
    calendar.setfirstweekday(calendar.SUNDAY)
    for i, j in enumerate(calendar.monthcalendar(date_in.year,date_in.month)):
        if calendar.monthcalendar(date_in.year,date_in.month)[0][0] == 1:
            i += 1
        for k, l in enumerate(j):
            if date_in.day == l:
                #found the date week position in a month and the day position in a week
                return(i+1, k+1)
```

```python
def get_dates_list(start_date, end_date, day_intervals):
    start_date = datetime.datetime.strptime(start_date, "%Y%m%d").date()
    end_date =  datetime.datetime.strptime(end_date, "%Y%m%d").date()
    date_list = []
    if (end_date - start_date).days % day_intervals == 0:
        for day in range(int(round((end_date - start_date).days/day_intervals))):
            a_date = (start_date + datetime.timedelta(day_intervals-1))
            if a_date >= end_date:
                a_date = end_date

            date_list.append([start_date, a_date])
            
            start_date = a_date + datetime.timedelta(1)
    else:
        for day in range(int((end_date - start_date).days/day_intervals)+1):
            a_date = (start_date + datetime.timedelta(day_intervals-1))
            if a_date >= end_date:
                a_date = end_date

            date_list.append([start_date, a_date])
            start_date = a_date + datetime.timedelta(1)
    return date_list
```
